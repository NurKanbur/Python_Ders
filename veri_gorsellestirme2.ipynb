```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
#öğrenci sayısı
ogrenci_sayisi = 30

#şube 1'in ders notları
s1_not1 = [78,67,89,23]
s1_not2 = [25,29,80,90]
s1_not3 = [57,92,91,47]
s1_not4 = [10,99,73,88]

#şube 2'nin ders notları
s2_not1 = [71, 80, 86, 90]
s2_not2 = [78, 83, 88, 91]
s2_not3 = [68, 79, 84, 89]
s2_not4 = [75, 82, 87, 92]

#ortalama hesaplama
s1_not1_ort = sum(s1_not1)/ ogrenci_sayisi
s1_not2_ort = sum(s1_not2)/ ogrenci_sayisi
s1_not3_ort = sum(s1_not3)/ ogrenci_sayisi
s1_not4_ort = sum(s1_not4)/ ogrenci_sayisi

s2_not1_ort = sum(s2_not1)/ ogrenci_sayisi
s2_not2_ort = sum(s2_not2)/ ogrenci_sayisi
s2_not3_ort = sum(s2_not3)/ ogrenci_sayisi
s2_not4_ort = sum(s2_not4)/ ogrenci_sayisi


#bar grafiği için veri oluşturma
ders_ort_s1 = [s1_not1_ort,s1_not2_ort,s1_not3_ort,s1_not4_ort]
ders_ort_s2 = [s2_not1_ort,s2_not2_ort,s2_not3_ort,s2_not4_ort]
dersler = ['ders 1','ders 2','ders 3','ders 4']

x = np.arange(len(dersler)) # x ekseninde derslerin konumunu belirleme.
bar_genislik = 0.35

#bar grafiği çizme

plt.bar(x,ders_ort_s1,width=bar_genislik,label='Şube 1')
plt.bar(x + bar_genislik, ders_ort_s2,width=bar_genislik,label='Şube 2')
plt.xlabel('Dersler')
plt.ylabel('Ortalama notlar')
plt.title('her ders için 2 şubenin ortalama notları')
plt.xticks(x + bar_genislik/2,dersler)
plt.legend()

#grafiği gösterme
plt.show()
```

```python
np.random.seed(0) # rastgele sayı üretimi için basit bir başlangıç noktası belirleyelim.

notlar = np.random.randint(0,101,size=(4,4,30)) #4 sınıf, 4 sınav, 30 öğrenci
#sınıfın ortalamalarını alalım
sınav_ortalama = np.mean(notlar,axis=2) # her bir sınavın ortalaması için son eksen boyunca ortalama alınır.

#her sınavın bar grafiğini çizdirelim
fig,ax =plt.subplots(1,4,figsize=(20,5)) # 1 satır, 4 sütunlu bir subplot oluşturalım.
sınavlar = ["sınav 1","sınav 2","sınav 3","sınav 4"]

for i in range(4):
    ax[i].bar(np.arange(4),sınav_ortalama[i],color='skyblue') #her bir sınavın ortalamasını çizdirelim.
    ax[i].set_xticks(np.arange(4)) # x ekseni için etiketleri belirleme
    ax[i].set_xticklabels(["şube 1","şube 2","şube 3","şube 4",]) # x ekseni etiketleri belirleme.
    ax[i].set_title(sınavlar[i]) #grafiğin başlığını belirleme.

plt.tight_layout() #grafiği sıkıştırma
plt.show() # grafiği gösterme

```

```python
import numpy as np
import matplotlib.pyplot as plt

# Verileri oluşturma
notlar = np.random.randint(0, 100, (30, 4))  # 4 sınıfın 30 öğrencisinin notları

# Her sınıfın ortalamasını hesapla
ortalama_notlar = np.mean(notlar, axis=0)

# Sınıf etiketleri
siniflar = ['A', 'B', 'C', 'D']

# Renkler
renkler = ['blue', 'green', 'red', 'orange']

# Bar plotları yan yana çizme
bars = plt.bar(siniflar, ortalama_notlar, color=renkler)

plt.xlabel('Sınıflar')
plt.ylabel('Ortalama Notlar')
plt.title('Sınıf Ortalama Notları')

# Legend (Leyenda)
plt.legend(bars, siniflar)

for i in range(len(siniflar)):
    plt.text(i,ortalama_notlar[i],f"{ortalama_notlar[i]:.2f}",ha="center",va="bottom")
plt.show()
```
