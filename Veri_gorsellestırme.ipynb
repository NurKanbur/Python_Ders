```python
deneme_3 = np.random.randint(0,100,(12))
deneme_4 = np.random.randint(0,100,(12))

x_eksen = np.arange(deneme_3.shape[0])
new_x_tic = ["Ocak","Şubat","Mart","Nisan","Mayıs","Haziran","Temmuz","Ağustos","Eylül","Ekim","Kasım","Aralık"]

plt.plot(x_eksen,deneme_3,linestyle="-.",color="blue",linewidth=2,label ="Deneme 3")
plt.plot(x_eksen,deneme_4,linestyle="--",color="red",linewidth=2,label = "Deneme 4")
plt.xticks(x_eksen,new_x_tic,rotation=45, ha ='right')
plt.xlabel("Aylar")
plt.ylabel("Notlar")
plt.title("2 öğrencinin 1 yıldaki notları")
plt.legend(loc="best")
plt.show()
```

```python
deneme_5 = np.random.randint(0,100,(30))
deneme_6 = np.random.randint(0,100,(30))
x_eksen = np.arange(deneme_5.shape[0])

plt.bar(x_eksen,deneme_5,color="orange", label="değer 5")
plt.bar(x_eksen,deneme_6,color="green", label="değer 6")
plt.legend(loc="best")
plt.show()
```

```python
#ortalama değer üzerinden karşılaştırma yapmaya yarar.
#özellikle de yıllar itibarı ile.
```

```python
deneme_5 = np.random.randint(0,100,(30))
deneme_6 = np.random.randint(0,100,(30))

ort_deneme_5 = np.mean(deneme_5)
ort_deneme_6 = np.mean(deneme_6)
x_uzunluk = ["Değer 5","Değer 6"]

ort_notlar = [ort_deneme_5,ort_deneme_6]

x_eksen = np.arange(deneme_5.shape[0])

plt.bar(x_uzunluk,ort_notlar,color=["orange","green"], label=["değer 5","değer 6"])

for i in range(len(x_uzunluk)):
    plt.text(i,ort_notlar[i], f"{ort_notlar[i]:.2f}",ha = "center", va = "bottom" )

plt.legend(loc="best")
plt.show()
```

```python
dizi = [60,40]
labels= ["60","40"]
colors =["blue","red"]
plt.pie(dizi,labels=labels,colors=colors,wedgeprops={'edgecolor':"black"})
plt.show
```

```python
dizi = [25,67,34,60,40]
#değerlerin hangi isimle gösterileceğini belirler.
labels= ["YBS","BST","Muhasebe","Gümrük","İşletme"]
#değerlerin gösterileceği dilimlerin rengini belirler.
colors =["blue","red","green","yellow","purple","pink"]
#öne çıkmasını istediğimiz değerin index numarasını 0.1 veya ne kadar öne çıkmasını istersek ona göre değiştiririz.
explode = [0.1,0,0,0,0]
#son olarak bu değişiklikleri pasta grafiğimiz için tanımlıyoruz.
#"%1.1f%%" ifadesi yüzdelik değerlerini göstermek için kullanılır.
plt.pie(dizi,labels=labels,colors=colors,explode=explode,autopct='%1.1f%%',wedgeprops={'edgecolor':"black"})
plt.show
```
