```python
deneme_3 = np.random.randint(0,100,(12))
deneme_4 = np.random.randint(0,100,(12))

x_eksen = np.arange(deneme_3.shape[0])
new_x_tic = ["Ocak","ubat","Mart","Nisan","Mays","Haziran","Temmuz","Austos","Eyll","Ekim","Kasm","Aralk"]

plt.plot(x_eksen,deneme_3,linestyle="-.",color="blue",linewidth=2,label ="Deneme 3")
plt.plot(x_eksen,deneme_4,linestyle="--",color="red",linewidth=2,label = "Deneme 4")
plt.xticks(x_eksen,new_x_tic,rotation=45, ha ='right')
plt.xlabel("Aylar")
plt.ylabel("Notlar")
plt.title("2 rencinin 1 yldaki notlar")
plt.legend(loc="best")
plt.show()
```

```python
deneme_5 = np.random.randint(0,100,(30))
deneme_6 = np.random.randint(0,100,(30))
x_eksen = np.arange(deneme_5.shape[0])

plt.bar(x_eksen,deneme_5,color="orange", label="deer 5")
plt.bar(x_eksen,deneme_6,color="green", label="deer 6")
plt.legend(loc="best")
plt.show()
```

```python
#ortalama deer zerinden karlatrma yapmaya yarar.
#zellikle de yllar itibar ile.
```

```python
deneme_5 = np.random.randint(0,100,(30))
deneme_6 = np.random.randint(0,100,(30))

ort_deneme_5 = np.mean(deneme_5)
ort_deneme_6 = np.mean(deneme_6)
x_uzunluk = ["Deer 5","Deer 6"]

ort_notlar = [ort_deneme_5,ort_deneme_6]

x_eksen = np.arange(deneme_5.shape[0])

plt.bar(x_uzunluk,ort_notlar,color=["orange","green"], label=["deer 5","deer 6"])

for i in range(len(x_uzunluk)):
    plt.text(i,ort_notlar[i], f"{ort_notlar[i]:.2f}",ha = "center", va = "bottom" )

plt.legend(loc="best")
plt.show()
```

```python
dizi = [25,67,34,60,40]
#deerlerin hangi isimle gsterileceini belirler.
labels= ["YBS","BST","Muhasebe","Gmrk","letme"]
#deerlerin gsterilecei dilimlerin rengini belirler.
colors =["blue","red","green","yellow","purple","pink"]
#ne kmasn istediimiz deerin index numarasn 0.1 veya ne kadar ne kmasn istersek ona gre deitiririz.
explode = [0.1,0,0,0,0]
#son olarak bu deiiklikleri pasta grafiimiz iin tanmlyoruz.
#"%1.1f%%" ifadesi yzdelik deerlerini gstermek iin kullanlr.
plt.pie(dizi,labels=labels,colors=colors,explode=explode,autopct='%1.1f%%',wedgeprops={'edgecolor':"black"})
plt.show
```
